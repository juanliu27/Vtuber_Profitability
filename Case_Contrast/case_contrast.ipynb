#### case changed by model suggestions compared to original data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
profit_case=pd.read_csv('casedata.csv',encoding='utf-8')
profit_case.head()
```

```python
plt.title('live_case_contrast')
plt.xlabel('Time')  # x轴标题
plt.ylabel('profit')  # y轴标题

plt.plot(range(len(profit_case)), list(profit_case['current']), label='real', color='darkred',marker='o',markerfacecolor='darkred',markersize='5')
plt.plot(range(len(profit_case)), list(profit_case['case1']), label='model', color='steelblue',marker='s',markerfacecolor='steelblue',markersize='5')
plt.plot(range(len(profit_case)), list(profit_case['case2']), label='model', color='darkslateblue',marker='s',markerfacecolor='darkslateblue',markersize='5')
#plt.plot(range(len(profit_live_case)), list(profit_live_case['live_case3']), label='model', color='darkolivegreen',marker='s',markerfacecolor='darkolivegreen',markersize='5')

plt.legend(['current', 'case1','case2'])
plt.xlim(5,31)
```

```python
plt.title('video_case_contrast')
plt.xlabel('Time')  # x轴标题
plt.ylabel('profit')  # y轴标题

plt.plot(range(len(profit_case)), list(profit_case['current']), label='real', color='darkred',marker='o',markerfacecolor='darkred',markersize='5')
plt.plot(range(len(profit_case)), list(profit_case['case1_video']), label='model', color='steelblue',marker='s',markerfacecolor='steelblue',markersize='5')
plt.plot(range(len(profit_case)), list(profit_case['case2_video']), label='model', color='darkslateblue',marker='s',markerfacecolor='darkslateblue',markersize='5')
#plt.plot(range(len(profit_live_case)), list(profit_live_case['live_case3']), label='model', color='darkolivegreen',marker='s',markerfacecolor='darkolivegreen',markersize='5')

plt.legend(['current', 'case1','case2'])
plt.xlim(10,31)
plt.ylim(8000,14000)
```
